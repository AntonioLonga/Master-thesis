```python
def my_plot(acc_vec,acc_spk,acc_spk_pca,name="DATASET_NAME"):
    pca_n_components = [250,300,500]
    n_bits_spk_pca = [9,11,13,15,17]
    n_bits_spk = [2,3,4,5,6,8,10]
    n_bits_vec = [2,4,6,8,10,12,14,16,18]

    acc_vec = np.mean(acc_vec, axis = 1)
    acc_spk = np.mean(acc_spk, axis = 1)

    acc_spk_pca = acc_spk_pca[1:]


    a = []
    a.append(acc_spk_pca[0][0])
    a.append(acc_spk_pca[1][0])
    a.append(acc_spk_pca[2][0])
    a.append(acc_spk_pca[3][0])
    a.append(acc_spk_pca[4][0])
    a = np.mean(a,axis=1)

    b = []
    b.append(acc_spk_pca[0][1])
    b.append(acc_spk_pca[1][1])
    b.append(acc_spk_pca[2][1])
    b.append(acc_spk_pca[3][1])
    b.append(acc_spk_pca[4][1])
    b = np.mean(b,axis=1)

    c = []
    c.append(acc_spk_pca[0][2])
    c.append(acc_spk_pca[1][2])
    c.append(acc_spk_pca[2][2])
    c.append(acc_spk_pca[3][2])
    c.append(acc_spk_pca[4][2])
    c = np.mean(c,axis=1)
    
    plt.figure(figsize=(8,8))
    plt.title(name)
    plt.plot(n_bits_vec,acc_vec, c='b',label="Vec_RF")
    plt.scatter(n_bits_vec,acc_vec, c='b')
    plt.plot([0] + n_bits_spk,acc_spk, c='r',label="Spk_Vec")
    plt.scatter([0] + n_bits_spk,acc_spk, c='r')

    plt.plot(n_bits_spk_pca,a,label="Spk_Vec_PCA 250",c='y')
    plt.scatter(n_bits_spk_pca,a,c='y')
    plt.plot(n_bits_spk_pca,b,label="Spk_Vec_PCA 300",c='g')
    plt.scatter(n_bits_spk_pca,b,c='g')
    plt.plot(n_bits_spk_pca,c,label="Spk_Vec_PCA 500",c='m')
    plt.scatter(n_bits_spk_pca,c,c='m')
    plt.legend()
    plt.show()
```
